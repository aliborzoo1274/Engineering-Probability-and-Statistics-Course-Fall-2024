### 1.

#### Simulate the uniform distribution:

```r
X <- runif(1e6, min = 0, max = 1)
```

### 2.

#### Form $Y = -2\ln(X)$:

```r
Y <- -2 * log(X)
```

### 3.

#### Plot the PDF of Y next to exponential distribution:

```r
rate <- 0.5
E <- rexp(1e6, rate)

par(mfrow = c(1, 2))

hist(Y, breaks = 100, probability = TRUE, 
     main = "PDF of Y",
     xlab = "y",
     col = "lightblue")
lines(density(Y), col = "red", lwd = 2)

hist(E, breaks = 100, probability = TRUE,
     main = "PDF of Exponential",
     xlab = "e",
     col = "lightgreen")
lines(density(E), col = "blue", lwd = 2)
```

We see that the two graphs have become similar to each other.

### 4.

#### Simulate the uniform distributions:

```r
U1 <- runif(1e6, min = 0, max = 1)
U2 <- runif(1e6, min = 0, max = 1)
```

### 5.

#### Form $Z1$ and $Z2$:

```r
R <- sqrt(-2 * log(U1))
T <- 2 * pi * U2
Z1 <- R * cos(T)
Z2 <- R * sin(T)
```

### 6.

#### Plot Z1, Z2 and Standard normal distribution:

```r
n <- seq(-4, 4, length=1000)
N <- dnorm(n, mean=0, sd=1)

par(mfrow = c(1, 3))

hist(Z1, breaks = 100, probability = TRUE, 
     main = "PDF of Z1",
     xlab = "z1",
     col = "lightblue")
lines(density(Z1), col = "green", lwd = 2)

hist(Z2, breaks = 100, probability = TRUE,
     main = "PDF of Z2",
     xlab = "z2",
     col = "lightgreen")
lines(density(Z2), col = "blue", lwd = 2)

plot(n, N, type="l", lwd=2, col="red", 
     main="PDF of Standard Normal",
     xlab="n", ylab="Density")
```

We see that the three graphs have become similar to each other.
